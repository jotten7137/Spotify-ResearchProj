```python
# Dependencies
import json
import pprint
import requests
import sys
import pandas as pd
import time
import matplotlib.pyplot as plt

# Retrieve API key from config.py
from config import key_id
from config import key_secret

#project specific packages
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
#Spotipy Documentation: https://spotipy.readthedocs.io/en/latest/

#Client Credential Flow
#requests that do not require access to a user's private data.
#higher rate limit
client_credentials_manager = SpotifyClientCredentials(client_id=key_id,client_secret=key_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager) #spotify object

#generates playlists
playlists = sp.user_playlists('spotify')

#displays key,value pairs within dictionary playlists
for key, value in playlists.items() :
    print (key, value)
```

```python
#Spotipy Documentation: https://spotipy.readthedocs.io/en/latest/
#generates playlists based on user:spotify playlists
playlists = sp.user_playlists('spotify')
while playlists:
    for i, playlist in enumerate(playlists['items']):
        print("%4d %s %s" % (i + 1 + playlists['offset'], playlist['uri'],  playlist['name']))
    if playlists['next']:
        playlists = sp.next(playlists)
    else:
        playlists = None
```

```python
#Top 19 artists as of Nov. 26 2018
artists = ['Selena Gomez',
           'Ariana Grande',
           'Drake',
           'Khalid',
           'Imagin Dragons',
           'Calvin Harris',
           'J Balvin',
           'Post Malone',
           'DJ Snake',
           'Ozuna',
           'Dua Lipa',
           'XXXTentacion',
           'Bad Bunny',
           'Marshmello',
           'Ed Sheeran',
           'Swae Lee',
           'David Guetta',
           'Maroon 5',
           'Queen',
           'Nicki Minaj']
```

```python
#search artist data
artist_list=[]
search_type = 'artist'
limit = 50
count=0
#creates list of dictionary based on type.

for i in artists:
    search_query = artists[count]
    search = sp.search(q=search_query,limit=limit,offset=0,type=search_type,market=None)
    artist_list.append(search)
    count+=1
print(artist_list)
```

```python
# DF to csv to see what we are working with
artist_df = pd.DataFrame.from_dict(artist_list)
print(artist_df)
artist_df.to_csv('spotify_data2.csv')
```

```python
#Based on reverse engineered spotify object

#data clean and capture
artist_list_capture = []
count = 0
counter = 0
while artists:
    try:
        for i, artist in enumerate(artist_list[count]['artists']):
            for j in artist_list[count]['artists']['items']:
                if artist_list[count]['artists']['items']:
                    goose = {'Artist Name':artist_list[count]['artists']['items'][counter]['name'], 
                              'Popularity':artist_list[count]['artists']['items'][counter]['popularity'],
                              'Followers':artist_list[count]['artists']['items'][counter]['followers']['total']}                            
                    counter +=1
                    artist_list_capture.append(goose)
                else:
                    break
                    counter = 0
                    count = 0
            if artist_list[count]['artists']:
                count +=1
                counter = 0
                next
            else:
                artist_list[count]['artists'] = None
                count +=1
                break
            break
    except IndexError:
        #reached the end of the list - nothing but recursion from here
        break   
print (artist_list_capture)
```

```python
#pandas Dataframe
artist_df = pd.DataFrame.from_records(artist_list_capture)
artist_df
```

```python
#collect that bad boy
artist_df.to_csv(f'csv_capture{time.time()}.csv')
```
